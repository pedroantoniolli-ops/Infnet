```python
# ============================================
# PRÉ-PROCESSAMENTO
# ============================================

numeric_transformer = Pipeline(
    steps=[
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

Dividir o dataset em conjunto de treino (training set) e conjunto de teste (test set) é uma prática fundamental em machine learning supervisionado, pois permite avaliar corretamente a capacidade de generalização do modelo.

Em outras palavras, queremos verificar se o modelo funciona bem em dados que ele nunca viu antes.

Por exemplo, para evitar avaliacao otimista do modelo.

Se treinarmos e avaliarmos o modelo usando os mesmos dados, o resultado será artificialmente bom.

Isso ocorre porque o modelo pode memorizar padrões específicos do dataset, ao invés de aprender relações gerais.

Exemplo:
- modelo treinado com 1000 registros
- testado nos mesmos 1000 registros

Resultado possível:
- Accuracy = 99%

Mas ao aplicar em novos dados:

Esse problema é chamado de overfitting.

Alem disso, sse torna adequado para simular dados futuros (nunca vistos)

O conjunto de teste representa novas observações que apareceriam no mundo real.

Processo:
1. Treinar o modelo usando dados de treino
2. Aplicar o modelo em dados de teste

- Accuracy

Pipeline:

No contexto de machine learning com scikit-learn, um pipeline é uma estrutura que permite encadear várias etapas de processamento e modelagem em um único fluxo organizado.

Ou seja, ele define uma sequência automática de transformações nos dados até chegar ao modelo final.

Um pipeline conecta etapas como:
- pré-processamento dos dados
- transformação das features
- treinamento do modelo

Tudo executado de forma sequencial.
Assim, o pipeline garante que todas as etapas sejam executadas na mesma ordem sempre que o modelo for treinado ou usado para prever novos dados.

```python
# ============================================
# DIVISÃO TREINO / TESTE
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDON_STATE
)
```

A função de avaliação é o conjunto de métricas usadas para medir o desempenho de um modelo de machine learning após o treinamento. Ela permite verificar quão bem o modelo está fazendo previsões ao comparar os valores previstos com os valores reais da variável-alvo.

Neste projeto, a função de avaliação serve para analisar a qualidade dos modelos treinados, como:
- Perceptron (baseline)
- Árvore de decisão
- Random Forest ou SVM

E comparar qual deles apresenta melhor desempenho.

Após treinar um modelo, fazemos previsões:
- modelo → gera previsões

Essas previsões são comparadas com os valores reais do dataset de teste.
- valor real vs valor previsto

A função de avaliação calcula métricas que indicam o nível de acerto e erro do modelo.

```python
# ============================================
# FUNÇÃO DE AVALIAÇÃO
# ============================================

def evaluate_model(model, X_test, y_test):

    y_pred = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1-score:", f1_score(y_test, y_pred))

    print("\nRelatório completo")
    print(classification_report(y_test, y_pred))
```

O Perceptron no scikit-learn é um classificador linear supervisionado utilizado principalmente para problemas de classificação binária. Ele é um dos algoritmos mais clássicos da área de redes neurais artificiais e também pode ser visto como um modelo linear simples que aprende um hiperplano de separação entre classes.

O Perceptron aprende uma função linear que combina as features do dataset para produzir uma decisão de classe.

Se houver apenas duas variáveis, o modelo aprende uma linha separando as classes.

Se houver três variáveis, aprende um plano.

Com muitas variáveis, temos um hiperplano em espaço multidimensional.

O objetivo do treinamento é ajustar os pesos 𝑤 para minimizar erros de classificação.

O algoritmo segue um processo iterativo:

1. Inicializa os pesos aleatoriamente.
2. Analisa uma observação.
3. Faz uma previsão.
4. Se a previsão estiver errada, ajusta os pesos.

Algumas vantagens do Perceptrom
- Simples
- Rapido
- Altamente interpretavel
- bom como baseline de modelos lineares

```python
# ============================================================
# MODELO BASELINE — PERCEPTRON
# ============================================================

print("\n==============================")
print("MODELO BASELINE — PERCEPTRON")
print("==============================")

perceptron_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", Perceptron(random_state=RANDON_STATE))
    ]
)

perceptron_pipeline.fit(X_train, y_train)

evaluate_model(perceptron_pipeline, X_test, y_test)

# coeficientes do perceptron
model = perceptron_pipeline.named_steps["model"]

print("\nCoeficientes do modelo:")
print(model.coef_)

print("\nBias (intercepto):")
print(model.intercept_)

```

Arvore de decisao:
A árvore de decisão é um modelo de machine learning supervisionado que realiza classificações ou previsões por meio de uma estrutura hierárquica de decisões, semelhante a um fluxograma. Esse modelo divide os dados em subconjuntos cada vez mais específicos, com base nos valores das variáveis (features), até chegar a uma decisão final.

Ela é muito utilizada porque é fácil de interpretar, explica claramente as regras de decisão e consegue capturar relações não lineares entre variáveis.

Uma árvore de decisão possui três elementos principais:

1. Nó raiz (root node)

É o primeiro ponto da árvore, onde ocorre a primeira divisão dos dados com base em uma variável considerada mais relevante para a previsão.

2. Nós internos (decision nodes)

São pontos intermediários da árvore que representam novas regras de decisão baseadas em outras variáveis.

Cada decisão divide o conjunto de dados em grupos menores.

3. Nós folha (leaf nodes)

São os resultados finais da classificação.

Para que a arvore possa aprender, durante o treinamento, o algoritmo procura dividir os dados de forma que as classes fiquem o mais separadas possível.

Ele escolhe a divisão que melhor separa os dados usando critérios matemáticos como:

- Gini Impurity: Mede o nível de mistura das classes em um nó. Quanto menor o valor, mais puro é o nó.

- Entropy (informação): Baseado na teoria da informação, mede o grau de desordem dos dados.

O objetivo é criar nós onde a maioria das observações pertença à mesma classe.

Alguns parâmetros controlam a complexidade da árvore.

- max_depth:	profundidade máxima da árvore
- min_samples_split	mínimo de observações para dividir um nó
- min_samples_leaf	mínimo de observações em um nó folha
- criterion	critério de divisão (gini ou entropy)

Esses parâmetros ajudam a evitar overfitting.

Vantagens da árvore de decisão
- fácil de interpretar
- gera regras claras
- captura relações não lineares
- funciona bem com dados categóricos e numéricos

Limitacoes
- Overfitting: Árvores muito profundas podem memorizar os dados de treino.
- Instabilidade: Pequenas mudanças nos dados podem gerar árvores diferentes.
- Menor desempenho isoladamente; Em muitos casos, modelos baseados em ensemble (Random Forest, Gradient Boosting) apresentam melhor desempenho.

Neste trabalho, a árvore de decisão será usada para:

Depois sera aplidado:

Para melhorar a árvore.

Em resumo, a árvore de decisão é um modelo de machine learning que aprende regras de decisão em forma de árvore para classificar dados. Ela divide o espaço das features em regiões menores, criando regras que levam da raiz até uma decisão final.

Esse modelo é particularmente útil por ser interpretável e capaz de capturar relações complexas entre variáveis.


```python
# ============================================================
# MODELO 2 — ÁRVORE DE DECISÃO
# ============================================================

print("\n==============================")
print("ÁRVORE DE DECISÃO")
print("==============================")

tree_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", DecisionTreeClassifier(random_state=RANDON_STATE))
    ]
)

tree_pipeline.fit(X_train, y_train)

evaluate_model(tree_pipeline, X_test, y_test)
```

A validação cruzada (cross-validation) combinada com Grid Search é uma técnica usada em machine learning para avaliar modelos de forma mais confiável e encontrar os melhores hiperparâmetros.

Essa etapa é muito importante porque ajuda a garantir que o modelo generalize bem para novos dados, evitando problemas como overfitting.

A validação cruzada consiste em dividir o conjunto de dados de treinamento em vários subconjuntos chamados folds.

Um método comum é o k-fold cross-validation.

```python
# ============================================================
# VALIDAÇÃO CRUZADA + GRID SEARCH
# ============================================================

print("\n==============================")
print("OTIMIZAÇÃO DA ÁRVORE")
print("==============================")

param_grid = {
    "model__max_depth": [3,5,10,None],
    "model__min_samples_split": [2,5,10],
    "model__min_samples_leaf": [1,2,5]
}

grid = GridSearchCV(
    tree_pipeline,
    param_grid,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Melhores parâmetros:")
print(grid.best_params_)

best_tree = grid.best_estimator_

evaluate_model(best_tree, X_test, y_test)
```

A Random Forest (Floresta Aleatória) é um algoritmo de machine learning supervisionado baseado em ensemble learning, ou seja, ele combina vários modelos simples para produzir uma previsão mais robusta e precisa.

Nesse caso, o modelo combina múltiplas árvores de decisão, formando uma “floresta”. Cada árvore faz uma previsão e o resultado final é obtido por votação entre as árvores.

Uma Random Forest funciona da seguinte maneira:
1. O algoritmo cria várias árvores de decisão.
2. Cada árvore é treinada com uma amostra aleatória dos dados.
3. Cada árvore utiliza um subconjunto aleatório de features.
4. As previsões de todas as árvores são combinadas.

Para classificação, o resultado final é a classe mais votada.

Por que usar varias arvores?

Uma única árvore de decisão pode ser instável e propensa a overfitting.

A Random Forest reduz esse problema porque:
- cada árvore vê dados diferentes
- cada árvore usa features diferentes
- o resultado final é a média das decisões

Isso torna o modelo:
- mais estável
- mais robusto
- mais preciso

```python

# ============================================================
# MODELO AVANÇADO — RANDOM FOREST
# ============================================================

print("\n==============================")
print("RANDOM FOREST")
print("==============================")

rf_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", RandomForestClassifier(random_state=RANDON_STATE))
    ]
)

rf_params = {
    "model__n_estimators": [100,200],
    "model__max_depth": [None,10,20],
    "model__min_samples_split": [2,5]
}

rf_grid = GridSearchCV(
    rf_pipeline,
    rf_params,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)

print("Melhores parâmetros Random Forest:")
print(rf_grid.best_params_)

best_rf = rf_grid.best_estimator_

evaluate_model(best_rf, X_test, y_test)
```

Cada algoritmo possui características diferentes e pode ter desempenhos distintos dependendo da natureza dos dados.

é possível identificar qual modelo produz previsões mais corretas.

```python


# ============================================================
# COMPARAÇÃO DE MODELOS
# ============================================================

print("\n==============================")
print("COMPARAÇÃO FINAL")
print("==============================")

models = {
    "Perceptron": perceptron_pipeline,
    "Decision Tree": tree_pipeline,
    "Optimized Tree": best_tree,
    "Random Forest": best_rf
}

for name, model in models.items():

    scores = cross_val_score(
        model,
        X,
        y,
        cv=5,
        scoring="f1"
    )

    print(name)
    print("F1 médio:", scores.mean())
    print("Desvio padrão:", scores.std())
    print()
```

O F1-score médio representa o desempenho médio do modelo ao longo dos folds da validação cruzada.

O F1-score é uma métrica que combina:
- Precision → qualidade das previsões positivas
- Recall → capacidade de encontrar casos positivos

Ele é muito utilizado quando:
- há desbalanceamento de classes
- queremos equilíbrio entre falsos positivos e falsos negativos

No caso do churn, isso é importante porque queremos identificar clientes que irão cancelar sem gerar muitos alarmes falsos.

O desvio padrão mede a variação do desempenho entre os folds da validação cruzada.

Interpretação:
- baixo desvio padrão → modelo estável
- alto desvio padrão → modelo instável

Isso indica consistência do modelo ao generalizar para diferentes subconjuntos dos dados.

- F1 médio: 0.446
- Desvio padrão: 0.163

O Perceptron apresentou o pior desempenho entre os modelos avaliados. Isso acontece porque:
- ele é um classificador linear
- o problema de churn provavelmente possui relações não lineares entre variáveis

Outro ponto importante:

O desvio padrão é alto (0.16), indicando que o desempenho varia bastante entre os folds.

Isso sugere que:
- o modelo é instável
- tem dificuldade em generalizar para diferentes subconjuntos de dados.

Dessa forma, o Perceptron funciona apenas como baseline, confirmando que modelos lineares simples não são suficientes para capturar os padrões do problema.
